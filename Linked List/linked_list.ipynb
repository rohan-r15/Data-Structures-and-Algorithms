```python
class Node:
    
    def __init__(self,data=None, next=None):
        self.data = data
        self.next = next
        
```

```python
class LinkedList:
    
    def __init__(self):
        self.head = None
        
        
    def insert_at_beginning(self, data):
        node = Node(data, self.head)
        self.head = node
        
      ## My way of insert at end  
#     def insert_at_end(self,data):
#         if self.head is None:
#             self.head = Node(data,None)
#             return
            
#         node = Node(data,None)
#         itr = self.head
        
#         while itr:
#             if itr.next == None:
#                 itr.next = node
#                 break
#             itr = itr.next
            
    
    ## from codebasics
    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data,None)
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        
        itr.next = Node(data, None)
        
        
    def insert_values(self,data_list):
        self.head = None
        for value in data_list:
            self.insert_at_end(value)
            
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count +=1
            itr = itr.next
        return count
    
    
    def remove_at(self, index):
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index")
            
        if index == 0:
            self.head = self.head.next
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == (index - 1):
                itr.next = itr.next.next 
                return
            
            itr = itr.next
            count += 1
        
            
    def insert_at(self,index,value):
        if index < 0 or index > self.get_length():
            raise Exception("Invalid Index")
        
        if index == 0:
            self.insert_at_beginning(value)
            return
        
        if index == self.get_length():
            self.insert_at_end(value)
            return
        
        itr = self.head
        count = 0
        while itr:
            if count == (index -1):
                itr.next = Node(value,itr.next.next)
                return
            
            itr = itr.next
            count += 1
        
                       
    def print(self): # print is a utility function 
        if self.head == None:
            print('Linked List is empty')
        
        llstr = ''
        itr = self.head
        
        while itr:
            llstr +=str(itr.data) + '--->'
            itr = itr.next
            
        print(llstr)
        
```

```python
ll = LinkedList()
ll.insert_at_beginning(20)
ll.insert_at_beginning(96)
ll.insert_at_end(190)
ll.insert_at_end(290)
ll.insert_at_beginning(0)
ll.print()
```
